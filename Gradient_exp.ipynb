```
import tensorflow as tf
def f(x, y):
    return (x**2 - 3*x+y**2)
def f_min():
    return (x**2 - 3*x+y**2)
def initialize():	
	x = tf.Variable(10.0) 
	y = tf.Variable(10.0) 
	return x, y
```

```
x, y = initialize()
lr = 0.1
for i in range(50):	
	with tf.GradientTape() as tape:
		z = f(x, y)
	grads = tape.gradient(z, [x, y])	
	x.assign(x - lr*grads[0].numpy())
	y.assign(y - lr*grads[1].numpy())
	print('z = {:.1f}, x = {:.1f}, y = {:.1f}'.format(f(x, y).numpy(), x.numpy(), y.numpy()))
```

```
x, y = initialize()
lr = 0.1
opt = tf.keras.optimizers.SGD(learning_rate=lr)
for i in range(50):
	with tf.GradientTape() as tape:
		z = f(x, y)
	grads = tape.gradient(z, [x, y])	
	processed_grads = [g for g in grads]
	grads_and_vars = zip(processed_grads, [x, y])
	opt.apply_gradients(grads_and_vars)
	print('z = {:.1f}, x = {:.1f}, y = {:.1f}'.format(f(x, y).numpy(), x.numpy(), y.numpy())) 
```

```
x, y = initialize()
opt = tf.keras.optimizers.SGD(learning_rate=lr)
for i in range(50):
	
	opt.minimize(f_min, var_list=[x, y])
	print ('z = {:.1f}, x = {:.1f}, y = {:.1f}'.format(f(x, y).numpy(), x.numpy(), y.numpy()))
```

```
obj_vals = []
coords = []
x, y = initialize()
opt = tf.keras.optimizers.SGD(learning_rate=lr)
for i in range(50):
	print ('z = {:.1f}, x = {:.1f}, y = {:.1f}'.format(f(x, y).numpy(), x.numpy(), y.numpy()))
	obj_vals.append(f(x, y).numpy())
	coords.append((x.numpy(), y.numpy()))
	opt.minimize(f_min, var_list=[x, y])
```

```
def plot_res(obj_vals, coords):
    plt.figure(figsize=(16, 6))
    plt.subplot(121)
    plt.contourf(X, Y, Z, 60, cmap=plt.cm.jet_r)
    plt.xlabel('x')
    plt.ylabel('y')
    cbar = plt.colorbar()


    xcoord = [x[0] for x in coords]
    ycoord = [x[1] for x in coords]
    plt.plot(xcoord, ycoord, '.-')
    plt.plot(xcoord[-1], ycoord[-1], "y*", markersize=12)

    plt.subplot(122)
    plt.plot(obj_vals, '.-')
    plt.plot([len(obj_vals) - 1], obj_vals[-1], "y*", markersize=12)
    plt.xlabel('Step')
    plt.ylabel('Objective')
    plt.show()
plot_res(obj_vals, coords)
```
